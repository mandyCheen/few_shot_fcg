```python
DATA_PATH = "/home/manying/Data/malware/Malware202403/fcg_ghidra/X86_64"
CSV_PATH = "/home/manying/Documents/malwareReport/Malware202403_info.csv"
```

```python
import pandas as pd
import os

df = pd.read_csv(CSV_PATH)
cpu = "Advanced Micro Devices X86-64"
df = df[df['CPU'] == cpu]

data_list = os.listdir(DATA_PATH)
```

```python
print(len(data_list))
print(len(df))
```

```python
new_df = df[df['file_name'].isin(data_list)]
print(len(new_df))
```

```python
FEW_SHOT_DATA_PATH = "/home/manying/Projects/fcgFewShot/dataset/data_ghidra_fcg/Advanced Micro Devices X86-64"
few_shot_family_list = os.listdir(FEW_SHOT_DATA_PATH)
print(len(few_shot_family_list))
```

```python
ori_family_list = new_df['family'].unique()
print(len(ori_family_list))
print(ori_family_list)
```

```python
openset_family_list = list(set(ori_family_list) - set(few_shot_family_list))
print(len(openset_family_list))
```

```python
openset_df = new_df[new_df['family'].isin(openset_family_list)]
print(len(openset_df))
value_counts = openset_df['family'].value_counts()
print(value_counts[value_counts > 20])
```

```python
dup_df = pd.read_csv("/home/manying/Documents/malwareReport/ghidra_fcg_temp_dup.csv")
print(dup_df.columns)
dup_df = dup_df[dup_df['file_name'].isin(data_list)]

openset_dup_df = dup_df[dup_df['family'].isin(openset_family_list)]
print(len(openset_dup_df))
```

```python
openset_dup_df.head()
openset_dup_df['dup_case_index'] = openset_dup_df['dup_case_index'].fillna(-1).astype(int)
```

```python
## filter out the packed samples

openset_dup_df = openset_dup_df[openset_dup_df['is_packed'] == False]
```

```python
print(len(openset_dup_df))
print(openset_dup_df['family'].value_counts())
```

```python
case_list = openset_dup_df['dup_case_index'].unique()
print(case_list)
case_list = case_list[case_list != -1]
```

```python
from tqdm import tqdm
print(len(case_list))
counts = openset_dup_df.value_counts('dup_case_index')
print(counts)
```

```python
for case_index in tqdm(case_list):
    case_df = openset_dup_df[openset_dup_df['dup_case_index'] == case_index]
    print(case_df['family'].value_counts())
    print(case_df[['file_name', 'family', 'dup_case_index']])
    print("")
```
